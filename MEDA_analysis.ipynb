```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import chart_studio.plotly as ply
```

```python
path = 'data/MagData-2025-11-13 21-39-01.csv'
data = pd.read_csv(path)

sample_rate = int(4)

#times = data['Times (s)'].to_numpy()
#times = times - times[0]
times = np.arange(len(data['Times (s)']))
B_x = data['B_x (nT)'].to_numpy()
B_y = data['B_y (nT)'].to_numpy()
B_z = data['B_z (nT)'].to_numpy()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x= times,
    y= B_x, 
    mode='lines', 
    name='B_x',
    line=dict(color='tomato', width=2)
))
fig.add_trace(go.Scatter(
    x= times,
    y= B_y, 
    mode='lines', 
    name='B_y',
    line=dict(color='rebeccapurple', width=2)
))
fig.add_trace(go.Scatter(
    x= times,
    y= B_z, 
    mode='lines', 
    name='B_z',
    line=dict(color='royalblue', width=2)
))
fig.update_layout(
    title='Triaxial Magnetometer Readings',
    xaxis_title='Time (s)',
    yaxis_title='Field Strength (nT)',
    template='plotly_dark',
    hovermode='x unified'
)

fig.show()
```
