```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import chart_studio.plotly as ply
```

```python
path = 'data/MagData-2025-12-17 20-21-30.csv'
data = pd.read_csv(path)

sample_rate = int(4)

#times = data['Times (s)'].to_numpy()
#times = times - times[0]
times = np.arange(len(data['Times (s)']))

B_x_raw = data['B_x (nT)'].to_numpy()
B_y_raw = data['B_y (nT)'].to_numpy()
B_z_raw = data['B_z (nT)'].to_numpy()

# B_x = B_x_raw - np.mean(B_x_raw)
# B_y = B_y_raw - np.mean(B_y_raw)
# B_z = B_z_raw - np.mean(B_z_raw)

B_x = B_x_raw - B_x_raw[0]
B_y = B_y_raw - B_y_raw[0]
B_z = B_z_raw - B_z_raw[0]


```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x= times,
    y= B_x, 
    mode='lines', 
    name='B_x',
    line=dict(color='tomato', width=2)
))
fig.add_trace(go.Scatter(
    x= times,
    y= B_y + times, 
    mode='lines', 
    name='B_y',
    line=dict(color='rebeccapurple', width=2)
))
fig.add_trace(go.Scatter(
    x= times,
    y= B_z, 
    mode='lines', 
    name='B_z',
    line=dict(color='royalblue', width=2)
))
fig.update_layout(
    title='Triaxial Magnetometer Readings',
    xaxis_title='Time (s)',
    yaxis_title='Field Strength (nT)',
    template='plotly_dark',
    hovermode='x unified'
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x= times,
    y= np.sqrt(B_x**2 + B_y**2 + B_z**2), 
    mode='lines', 
    name='B_x',
    line=dict(color='tomato', width=2)
))
fig.update_layout(
    title='Triaxial Magnetometer Readings',
    xaxis_title='Time (s)',
    yaxis_title='Field Strength (nT)',
    template='plotly_dark',
    hovermode='x unified'
)

fig.show()
```

```python
ch1_fft = np.fft.rfft(B_y)
ch1_fftfreq = np.fft.rfftfreq(len(B_y),1./sample_rate)

fig = go.Figure(data=go.Scatter(
    x= ch1_fftfreq,
    y= np.log(np.abs(ch1_fft)), 
    mode='lines', 
    name='Channel 1',
    line=dict(color='rebeccapurple', width=2)
))
fig.update_layout(
    title='Secondary Coil Frequency Spectrum (Channel 1)',
    xaxis_title='Frequency (Hz)',
    yaxis_title='Power',
    template='plotly_dark',
    hovermode='x unified'
)
fig.show()
```
